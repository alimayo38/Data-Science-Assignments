```python

G = nx.Graph()
G.add_nodes_from(['a', 'b', 'c', 'd', 'e', 'f', 'g'])
edges = [
    ('a', 'c'), ('a', 'e'), ('a', 'g'), ('b', 'b'), ('b', 'f'), ('c', 'a'),
    ('c', 'f'), ('c', 'g'), ('d', 'e'), ('e', 'd'), ('f', 'b'), ('f', 'c'),('f','g'),('g','a'),('g','c'),('g','f')
]

G.add_edges_from(edges)
for node in G.nodes():
  if node in G.neighbors(node):
    print(f"Loop detected at node {node}")


nx.draw(G, with_labels=True)
```

```python
import random

for u, v in G.edges():
    G[u][v]['weight'] = random.randint(1, 10)


shortest_path = nx.dijkstra_path(G, 'a', 'b')
shortest_path_length = nx.dijkstra_path_length(G, 'a', 'b')


print("Shortest path:", shortest_path)
print("Length of shortest path:", shortest_path_length)

nx.draw(G, with_labels=True)
```

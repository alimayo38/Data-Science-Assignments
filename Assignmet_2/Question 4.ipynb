```python
#import Google Drive and mount the entire drive

from google.colab import drive
drive.mount('/content/drive')
#load the dataset (csv file) into the Pandas dataframe

df_nuclear_wastes= pd.read_csv('/content/drive/My Drive/DS-Datasets/nuclear_waste_sites.csv')
```

```python

center_coords = [39.8283, -98.5795]

map_us = folium.Map(location=center_coords, zoom_start=4, tiles='OpenStreetMap')

for _, row in df_nuclear_wastes.iterrows():
    folium.Marker(location=[row['lat'], row['lon']], popup=row['text']).add_to(map_us)


map_us
```

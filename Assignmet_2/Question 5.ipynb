```python
#import Google Drive and mount the entire drive

from google.colab import drive
drive.mount('/content/drive')
#load the dataset (csv file) into the Pandas dataframe

file_path = '/content/drive/My Drive/DS-Datasets/pak-heritage-sites.csv'
df_pak_heritage_sites = pd.read_csv(file_path, header=None)
```

```python

df_pak_heritage_sites.columns = ['Latitude', 'Longitude', 'Name']

# Coordinates of Pakistan to center the map
pakistan_coords = [30.3753, 69.3451]

# Create base map
pak_map = folium.Map(location=pakistan_coords, zoom_start=5)


marker_cluster = MarkerCluster().add_to(pak_map)


for index, row in df_pak_heritage_sites.iterrows():
    folium.Marker(location=[row['Latitude'], row['Longitude']], popup=row['Name']).add_to(marker_cluster)

pak_map
```

```python
#import Google Drive and mount the entire drive

from google.colab import drive
drive.mount('/content/drive')
#load the dataset (csv file) into the Pandas dataframe

df_diamonds= pd.read_csv('/content/drive/My Drive/DS-Datasets/diamonds.csv')
```

```python
# Filter the dataset
filtered_diamonds = df_diamonds[(df_diamonds['clarity'] == 'SI2') & (df_diamonds['color'] == 'E')]

# Set color palette based on 'cut' categories
cut_colors = {'Fair': 'red', 'Good': 'orange', 'Very Good': 'yellow', 'Premium': 'green', 'Ideal': 'blue'}
sns.set_palette(cut_colors.values())

# Plot the relationship between 'carat' and 'price' with different colors for each 'cut'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=filtered_diamonds, x='carat', y='price', hue='cut', palette=cut_colors)
plt.title('Relationship between Carat and Price of Diamonds (SI2 Clarity, E Color)', fontsize=16)
plt.xlabel('Carat', fontsize=14)
plt.ylabel('Price', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='Cut', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#import Google Drive and mount the entire drive

from google.colab import drive
drive.mount('/content/drive')
#load the dataset (csv file) into the Pandas dataframe

df_population = pd.read_csv('/content/drive/My Drive/DS-Datasets/world_pop.csv')
```

```python
df_population.head(10)
```

```python
top_10_countries_2020 = df_population.nlargest(10, 'year_1995')
top_10_data = df_population[df_population['country'].isin(top_10_countries_2020['country'])]
pivot_data = top_10_data.set_index('country').T
plt.figure(figsize=(12, 8))
line_styles = ['-', '--', '-.', ':']
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', '#FFA500', '#800080', '#008080']

for i, country in enumerate(pivot_data.columns):
    plt.plot(pivot_data.index, pivot_data[country], label=country, linestyle=line_styles[i % len(line_styles)], color=colors[i])

plt.title('Population Comparison of Top 10 Most Populated Countries (1960-2020)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Population', fontsize=14)
plt.legend(loc='upper left', fontsize=10)
plt.grid(True)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.tight_layout()
plt.show()
```
